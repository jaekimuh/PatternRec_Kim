```python
import numpy as np
  

array = np.random.randint(10, size=(100, 100))

U, S, VT = np.linalg.svd(array,full_matrices=0)
S = np.diag(S)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [8, 4]

plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values')
plt.show()
```
