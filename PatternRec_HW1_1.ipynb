```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
from skimage import color


A = imread(os.path.join('..','python','DATA', 'dog.jpg'))
print(A.shape)
imgGray = color.rgb2gray(A)
img = plt.imshow(imgGray)
img.set_cmap('gray')
plt.axis('off')
plt.show()

```

```python
# Do SVD

U, S, VT = np.linalg.svd(imgGray, full_matrices=False)
S = np.diag(S)
r = 5
U1 = U[:,:r]
print(U1.shape)
UT = U1.transpose()
C = UT@U1
print(C.shape)
#print(C)

for r in (5, 20, 25):
    Ur = U[:,:r]
    print(Ur.shape)
    UT = np.transpose(Ur)
    UTU = UT@Ur
    err = np.linalg.norm(UTU - np.identity(r))
    print(err)


```
